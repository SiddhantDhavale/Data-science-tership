# box plot

```python
marks=[12,23,34,45,56,67,78,89]
```

```python
plt.boxplot(marks);
```

```python
q3=np.percentile(marks,75)
q1=np.percentile(marks,25)
```

```python
q3,q1
```

```python
iqr=q3-q1
iqr #interquartile range
```

```python
#lower bound
lb=q1-1.5*iqr
#upper bound
up=q3+1.5*iqr
```

```python
lb,up
```

```python
tips=sns.load_dataset('tips')
```

```python
import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)
```

```python
sns.jointplot(x='tip',y='total_bill',data=tips,hue='time');
plt.show()
```

```python
#histogram
plt.xlabel('Tip Amount')
plt.ylabel('Frequency')
sns.histplot(tips['tip'],kde=True)

plt.show()
```
